```python
#importing the Libraries#
import pandas as pd
```

```python
#Reading the Dataset#
dataset=pd.read_csv('50_Startups.csv')
```

```python
#what saved in this file#
dataset
```

```python
#create dummy data set. #drop a column#
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#what saved in this file#
dataset
```

```python
#catogorical data convert to numerical data#
dataset=dataset.astype(int)
```

```python
#what saved in this file#
dataset
```

```python
#show all columns in dataset#
dataset.columns
```

```python
#independent variable means input.  #split input variables. #dataset assigned to independent variable#
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#what saved in this file#
independent
```

```python
#dependent variable means output.  #split output variables. #dataset assigned to dependent variable#
dependent=dataset[['Profit']]
```

```python
#what saved in this file#
dependent
```

```python
#machine learning algorithms provider is sklearn #split train set and test set#
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
```

```python
X_train
```

```python
#Model creation#
from sklearn.svm import SVR
regressor=SVR(kernel='rbf',C=1000)
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
#cross check prediction#
y_pred=regressor.predict(X_test)
```

```python
#find r2_score#
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#show this file#
r_score
```

```python
import pickle
filename="50_Startups.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
